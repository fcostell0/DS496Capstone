# EDA Part 2: Census Education Data
Examining the data from the 2013-2023 ACS.

### Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading Data

```python
educationData = pd.read_csv('C:/Users/finco/Documents/GitHub/DS496Capstone/Processed Data/educationData.csv')
```

### Plot Settings

```python
plt.rcParams['font.family'] = 'Times New Roman'
sns.set_palette("crest")
```

### Summary Statistics

```python
educationData.describe()
```

Out of all the education levels, HS Graduate tends to have a plurality in our congressional districts.

### Median Earnings

```python
sns.histplot(data=educationData['MedianEarnings'], bins=25, linewidth = 0.5)
plt.title('Median Earnings of Congressional Districts')
plt.xlabel('Median Age')
plt.show()
```

Unimodal around 50k, should note the skew off to the right. The extremely high income congressional districts are likely Democratic. 

### Less Than HS Education

```python
sns.histplot(data=educationData['Education<9th'], bins=25, linewidth = 0.5)
plt.title('Less than HS Education % by Congressional District')
plt.xlabel('Less than HS Education %')
plt.show()
```

Note the right skew, likely highly Republican districts.

### Some HS Education

```python
sns.histplot(data=educationData['HSNoDiploma'], bins=25, linewidth = 0.5)
plt.title('Some HS Education % by Congressional District')
plt.xlabel('Some HS Education %')
plt.show()
```

Interesting how this is a much more even distribution.

### HS Graduate

```python
sns.histplot(data=educationData['HSGraduate'], bins=25, linewidth = 0.5)
plt.title('HS Graduate % by Congressional District')
plt.xlabel('HS Graduate %')
plt.show()
```

Probably will be the most even distribution of the data.

### Some College

```python
sns.histplot(data=educationData['SomeCollege'], bins=25, linewidth = 0.5)
plt.title('Some College % by Congressional District')
plt.xlabel('Some College %')
plt.show()
```

I suspect that left tail is the result of low-education districts and ultra-high education districts where most college students graduate.

### Associates Degree

```python
sns.histplot(data=educationData['Associates'], bins=25, linewidth = 0.5)
plt.title('Associates % by Congressional District')
plt.xlabel('Associates %')
plt.show()
```

That spike at 16 is interesting, likely districts with many trade workers.

### Bachelors Degree

```python
sns.histplot(data=educationData['Bachelors'], bins=25, linewidth = 0.5)
plt.title('Bachelors Degree % by Congressional District')
plt.xlabel('Bachelors Degree %')
plt.show()
```

Probably the most important chart of all our EDA. I'm a little suprised so many districts have <20% of their population as college graduates. Given recent trends, a higher percentage here is much worse for Republicans.

### Graduate Degree

```python
sns.histplot(data=educationData['Graduate'], bins=25, linewidth = 0.5)
plt.title('Graduate Degree % by Congressional District')
plt.xlabel('Graduate Degree %')
plt.show()
```

No real suprises. That tail going over 25% are likely coastal districts with strong Democratic majorities.

### Correlation Heatmap

```python
corrMatrix = educationData[['Education<9th', 'HSNoDiploma', 'HSGraduate', 'SomeCollege', 'Associates', 'Bachelors', 'Graduate']].corr()

sns.heatmap(corrMatrix, annot=True, cmap='crest')
```

Some of these correlations are fairly high. We will definitely need to use PCA on this dataset.
