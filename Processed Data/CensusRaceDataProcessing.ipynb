## A Note on Redistricting
For the sake of future use, we would prefer to use the ACS surveys conducted the year *before* each congressional election, as their data is released in the following Fall. Unfortunately, our data covers 2 elections after redistricting cycles, 2012 and 2022. The 2011 and 2021 ACS surveys *do not* take redistricting into account, so we will have to use the 2012 and 2022 surveys for the sake of accuracy.

### Dropping Columns
We can borrow our functions from the education data.

```python
def dropColumns12(dataframe):
    featureList = ['NAME', # CD Name
                    'DP05_0017E', # Median age
                    'DP05_0066PE', # % Hispanic or Latino
                    'DP05_0072PE', # % White, not hispanic
                    'DP05_0073PE', # % Black, not hispanic
                    'DP05_0074PE', # % Native american, not hispanic
                    'DP05_0075PE', # % Asian, not hispanic
                    ]

    featureDict = {
        'NAME': 'Congressional District',
        'DP05_0017E':'MedianAge',
        'DP05_0066PE':'Hispanic',
        'DP05_0072PE':'White',
        'DP05_0073PE':'Black',
        'DP05_0074PE':'NativeAmerican',
        'DP05_0075PE':'Asian'
    }

    dataframe = dataframe[featureList]
    dataframe = dataframe.rename(columns = featureDict)
    dataframe = dataframe.drop(index = 0)
    return dataframe

ACS2012 = dropColumns12(ACS2012)
ACS2012['year'] = 2012
ACS2012
```

### Text Modification
Again, we can borrow our other functions.

```python
def textProcessing(dataframe):
    dataframe['state'] = dataframe['Congressional District'].str.extract(r',\s([A-Za-z\s]+)$')

    state_abbr = {
        'Alabama': 'AL', 'Alaska':'AK', 'Arizona':'AZ', 'Arkansas':'AR', 'California':'CA', 'Colorado':'CO', 
        'Connecticut':'CT','Delaware':'DE','Florida':'FL','Georgia':'GA', 'Hawaii':'HI','Idaho':'ID','Illinois':'IL',
        'Indiana':'IN', 'Iowa':'IA', 'Kansas':'KS', 'Kentucky':'KY', 'Louisiana':'LA', 'Maine':'ME', 'Maryland':'MD',
        'Massachusetts':'MA', 'Michigan':'MI', 'Minnesota':'MN', 'Mississippi':'MS', 'Missouri':'MO', 'Montana':'MT',
        'Nebraska':'NE', 'Nevada':'NV', 'New Hampshire':'NH', 'New Jersey':'NJ', 'New Mexico':'NM','New York':'NY', 'North Carolina':'NC',
        'North Dakota':'ND', 'Ohio':'OH', 'Oklahoma':'OK', 'Oregon':'OR', 'Pennsylvania':'PA', 'Rhode Island':'RI', 
        'South Carolina':'SC', 'South Dakota':'SD', 'Tennessee':'TN', 'Texas':'TX', 'Utah':'UT', 'Vermont':'VT',
        'Virginia':'VA', 'Washington':'WA', 'West Virginia':'WV', 'Wisconsin':'WI', 'Wyoming':'WY'
    }

    dataframe['state_po'] = dataframe['state'].map(state_abbr)
    dataframe['district'] = dataframe['Congressional District'].str.extract(r'District (\d+)')

    dataframe = dataframe.dropna(subset=['state_po'])
    dataframe['district'] = dataframe['district'].fillna(0)

    dataframe = dataframe.drop(['state', 'Congressional District'], axis = 1)

    return dataframe

ACS2012 = textProcessing(ACS2012)
ACS2012
```

## 2017

## 2022
The feature codes changed again. Time to re-write dropColumns again.

```python
def dropColumns22(dataframe):
    featureList = ['NAME', # CD Name
                    'DP05_0018E', # Median age   
                    'DP05_0076PE', # % Hispanic or Latino    DIFFERENT  AGAIN
                    'DP05_0082PE', # % White, not hispanic    DIFFERENT  AGAIN
                    'DP05_0083PE', # % Black, not hispanic     DIFFERENT  AGAIN
                    'DP05_0084PE', # % Native american, not hispanic   DIFFERENT AGAIN
                    'DP05_0085PE', # % Asian, not hispanic    DIFFERENT  AGAIN
                    ]

    featureDict = {
        'NAME': 'Congressional District',
        'DP05_0018E':'MedianAge',
        'DP05_0076PE':'Hispanic',
        'DP05_0082PE':'White',
        'DP05_0083PE':'Black',
        'DP05_0084PE':'NativeAmerican',
        'DP05_0085PE':'Asian'
    }

    dataframe = dataframe[featureList]
    dataframe = dataframe.rename(columns = featureDict)
    dataframe = dataframe.drop(index = 0)
    return dataframe
```
