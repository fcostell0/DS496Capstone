### Dropping Columns
We only need a few of these columns, time to filter them out.

```python
def dropColumns1113(dataframe):
    featureList = ['NAME', # CD Name
                    'S1501_C01_006E', # Total Population > 25
                    'S1501_C01_007E', # Education < 9th grade
                    'S1501_C01_008E', # 9th-12th no diploma
                    'S1501_C01_009E', # High School graduate
                    'S1501_C01_010E', # Some college
                    'S1501_C01_011E', # Associates degree
                    'S1501_C01_012E', # Bachelor's degree
                    'S1501_C01_013E', # Graduate/Professional Degree
                    'S1501_C01_032E' # Median earnings
                    ]

    featureDict = {
        'NAME': 'Congressional District',
        'S1501_C01_006E' : 'TotalPop',
        'S1501_C01_007E' : 'Education<9th',
        'S1501_C01_008E': 'HSNoDiploma',
        'S1501_C01_009E': 'HSGraduate',
        'S1501_C01_010E': 'SomeCollege',
        'S1501_C01_011E' : 'Associates',
        'S1501_C01_012E' : 'Bachelors',
        'S1501_C01_013E' : 'Graduate',
        'S1501_C01_032E' : 'MedianEarnings'
    }

    dataframe = dataframe[featureList]
    dataframe = dataframe.rename(columns = featureDict)
    dataframe = dataframe.drop(index = 0)
    return dataframe

ACS2011 = dropColumns1113(ACS2011)
ACS2011['year'] = 2011
ACS2011
```

### Text Modification
Now we need to modify the text in 'Congressional District'.

```python
def textProcessing(dataframe):
    dataframe['state'] = dataframe['Congressional District'].str.extract(r',\s([A-Za-z\s]+)$')

    state_abbr = {
        'Alabama': 'AL', 'Alaska':'AK', 'Arizona':'AZ', 'Arkansas':'AR', 'California':'CA', 'Colorado':'CO', 
        'Connecticut':'CT','Delaware':'DE','Florida':'FL','Georgia':'GA', 'Hawaii':'HI','Idaho':'ID','Illinois':'IL',
        'Indiana':'IN', 'Iowa':'IA', 'Kansas':'KS', 'Kentucky':'KY', 'Louisiana':'LA', 'Maine':'ME', 'Maryland':'MD',
        'Massachusetts':'MA', 'Michigan':'MI', 'Minnesota':'MN', 'Mississippi':'MS', 'Missouri':'MO', 'Montana':'MT',
        'Nebraska':'NE', 'Nevada':'NV', 'New Hampshire':'NH', 'New Jersey':'NJ', 'New Mexico':'NM','New York':'NY', 'North Carolina':'NC',
        'North Dakota':'ND', 'Ohio':'OH', 'Oklahoma':'OK', 'Oregon':'OR', 'Pennsylvania':'PA', 'Rhode Island':'RI', 
        'South Carolina':'SC', 'South Dakota':'SD', 'Tennessee':'TN', 'Texas':'TX', 'Utah':'UT', 'Vermont':'VT',
        'Virginia':'VA', 'Washington':'WA', 'West Virginia':'WV', 'Wisconsin':'WI', 'Wyoming':'WY'
    }

    dataframe['state_po'] = dataframe['state'].map(state_abbr)
    dataframe['district'] = dataframe['Congressional District'].str.extract(r'District (\d+)')

    dataframe = dataframe.dropna(subset=['state_po'])
    dataframe['district'] = dataframe['district'].fillna(0)

    dataframe = dataframe.drop(['state', 'Congressional District'], axis = 1)

    return dataframe

ACS2011 = textProcessing(ACS2011)
ACS2011
```

## 2015
Discovered the column codes are different after 2013. Incredibly annoying, have to re-code the dropColumns function now

```python
def dropColumns15(dataframe):
    featureList = ['NAME', # CD Name
                    'S1501_C01_006E', # Total Population > 25  SAME AS 13
                    'S1501_C01_007E', # Education < 9th grade  SAME AS 13
                    'S1501_C01_008E', # 9th-12th no diploma  SAME AS 13
                    'S1501_C01_009E', # High School graduate  SAME AS 13
                    'S1501_C01_010E', # Some college SAME AS 13
                    'S1501_C01_011E', # Associates degree  SAME AS 13
                    'S1501_C01_012E', # Bachelor's degree SAME AS 13
                    'S1501_C01_013E', # Graduate/Professional Degree SAME AS 13
                    'S1501_C01_059E' # Median earnings    DIFFERENT
                    ]

    featureDict = {
        'NAME': 'Congressional District',
        'S1501_C01_006E' : 'TotalPop',
        'S1501_C01_007E' : 'Education<9th',
        'S1501_C01_008E': 'HSNoDiploma',
        'S1501_C01_009E': 'HSGraduate',
        'S1501_C01_010E': 'SomeCollege',
        'S1501_C01_011E' : 'Associates',
        'S1501_C01_012E' : 'Bachelors',
        'S1501_C01_013E' : 'Graduate',
        'S1501_C01_059E' : 'MedianEarnings'
    }

    dataframe = dataframe[featureList]
    dataframe = dataframe.rename(columns = featureDict)
    dataframe = dataframe.drop(index = 0)
    return dataframe
```

```python
ACS2015 = pd.read_csv('C:/Users/finco/Documents/GitHub/DS496Capstone/Raw Data/Education/ACSST1Y2015.S1501-Data.csv')
ACS2015 = dropColumns15(ACS2015)
ACS2015 = textProcessing(ACS2015)
ACS2015['year'] = 2015
ACS2015
```

Now we need to code a function to convert all the education statistics into percentages.

```python
def convertTotalsToPercentage(dataframe):
    lst = ['Education<9th', 'HSNoDiploma', 'HSGraduate', 'SomeCollege', 'Associates', 'Bachelors', 'Graduate']
    dataframe['TotalPop'] = pd.to_numeric(dataframe['TotalPop'])

    for column in lst:
        dataframe[column] = pd.to_numeric(dataframe[column])
        dataframe[column] = (dataframe[column] / dataframe['TotalPop']) * 100

    return dataframe

ACS2015 = convertTotalsToPercentage(ACS2015)
ACS2015
```

## Tying it all together

```python
educationData = pd.concat([ACS2011, ACS2013, ACS2015, ACS2017, ACS2019, ACS2021, ACS2023], axis = 0, ignore_index = True)
educationData = educationData.drop(['TotalPop'], axis = 1)
educationData[['Education<9th', 'HSNoDiploma', 'HSGraduate', 'SomeCollege', 'Associates', 'Bachelors', 'Graduate', 'MedianEarnings']] = educationData[['Education<9th', 'HSNoDiploma', 'HSGraduate', 'SomeCollege', 'Associates', 'Bachelors', 'Graduate', 'MedianEarnings']].apply(pd.to_numeric)
educationData = educationData.round(2)
educationData
```

### Adjusting Earnings to Inflation

```python
multiplier_dict = {
    2011: 1.44,
    2013: 1.38,
    2015: 1.36,
    2017: 1.31,
    2019: 1.26,
    2021: 1.21,
    2023: 1.06
}

educationData['MedianEarnings'] = educationData.apply(lambda row: row['MedianEarnings'] * multiplier_dict.get(row['year'], 1), axis = 1)
educationData
```

## UPDATE
I have realized that the 2011 ACS data and the 2021 ACS data doesn't account for their redistricting cycles. These need to be replaced with 2013 and 2023 respectively

```python
educationData = educationData[(educationData['year'] != 2011) & (educationData['year'] != 2021)]
educationData

replacementData = pd.concat([ACS2013, ACS2023], axis=0, ignore_index=True)
replacementData = replacementData.drop(['TotalPop'], axis = 1)
replacementData[['Education<9th', 'HSNoDiploma', 'HSGraduate', 'SomeCollege', 'Associates', 'Bachelors', 'Graduate', 'MedianEarnings']] = replacementData[['Education<9th', 'HSNoDiploma', 'HSGraduate', 'SomeCollege', 'Associates', 'Bachelors', 'Graduate', 'MedianEarnings']].apply(pd.to_numeric)
replacementData = replacementData.round(2)

replacementData['MedianEarnings'] = replacementData.apply(lambda row: row['MedianEarnings'] * multiplier_dict.get(row['year'], 1), axis = 1)

replacementData['year'] = replacementData['year'] - 1

educationData['year'] = educationData['year'] + 1

pd.concat([replacementData, educationData], axis=0, ignore_index=True)
```
