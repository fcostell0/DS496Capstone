# House Election Data

This dataset contains election results from house races from 1976 to 2022. We will only need data from 2012
 onwards.

```python
import pandas as pd
```

```python
rawData = pd.read_csv("C:/Users/finco/Documents/GitHub/DS496Capstone/Raw Data/Political/1976-2022-house.csv")
rawData
```

```python
rawData = rawData[rawData['year'] >= 2012]
rawData
```

```python
rawData[['stage', 'runoff', 'special']].value_counts() # Checking to see if there's a significant amount of runoff or special elections in dataset
```

### Column Dropping
We can drop many of these columns, as we only need to know if a Republican won their election or not. The fusion_ticket column deserves further investigation though.

```python
rawData = rawData.drop(['state', 'state_fips', 'state_cen', 'state_ic', 'office', 'stage', 'runoff', 'special', 'candidate', 'mode', 'unofficial', 'version', 'writein', 'fusion_ticket', 'totalvotes'], axis=1)
rawData
```

### "Winner" Label
We need to label the winners of each respective election.

```python
rawData['max_votes'] = rawData.groupby(['year', 'state_po', 'district'])['candidatevotes'].transform('max')
rawData = rawData[rawData['candidatevotes'] == rawData['max_votes']]

rawData['republican_victory'] = (rawData['party'] == 'REPUBLICAN')
rawData = rawData.drop(['party', 'candidatevotes', 'max_votes'], axis = 1)

rawData
```

### Merging with 2024 Data
Combining with the data that I put together myself.

```python
data2024 = pd.read_csv("C:/Users/finco/Documents/GitHub/DS496Capstone/Raw Data/Political/2024manualresults.csv")
data2024
```

```python
electionData = pd.concat([rawData, data2024])
electionData
```

### Adding Political Context Features

```python
electionData['republican_president'] = (electionData['year'] > 2017) & (electionData['year'] < 2021)
electionData['midterm'] = (electionData['year'] == 2014) | (electionData['year'] == 2018) | (electionData['year'] == 2022)
electionData
```

```python
electionData.to_csv('electionLabels.csv', index=False)
```
