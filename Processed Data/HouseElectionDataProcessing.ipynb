### Column Dropping
We can drop many of these columns, as we only need to know if a Republican won their election or not. The fusion_ticket column deserves further investigation though.

```python
rawData = rawData.drop(['state', 'state_fips', 'state_cen', 'state_ic', 'office', 'stage', 'runoff', 'special', 'candidate', 'mode', 'unofficial', 'version', 'writein', 'fusion_ticket', 'totalvotes'], axis=1)
rawData
```

### "Winner" Label
We need to label the winners of each respective election.

```python
rawData['max_votes'] = rawData.groupby(['year', 'state_po', 'district'])['candidatevotes'].transform('max')
rawData = rawData[rawData['candidatevotes'] == rawData['max_votes']]

rawData['republican_victory'] = (rawData['party'] == 'REPUBLICAN')
rawData = rawData.drop(['party', 'candidatevotes', 'max_votes'], axis = 1)

rawData
```

## Cook PVI Data

### 2012

```python
PVI2012 = pd.read_excel('C:/Users/finco/Documents/GitHub/DS496Capstone/Raw Data/Political/Cook PVI 1997-2025.xlsx', sheet_name='Post-2012 redistricting (2012)')
PVI2012
```

```python
def process_pvi(pvi):
    if 'D+' in pvi: 
        return -float(pvi.replace('D+', ''))  
    elif 'R+' in pvi: 
        return float(pvi.replace('R+', ''))
    return None 
```

```python
def processStateDistrict(dataframe, year):
    state_abbr = {
            'Alabama': 'AL', 'Alaska':'AK', 'Arizona':'AZ', 'Arkansas':'AR', 'California':'CA', 'Colorado':'CO', 
            'Connecticut':'CT','Delaware':'DE','Florida':'FL','Georgia':'GA', 'Hawaii':'HI','Idaho':'ID','Illinois':'IL',
            'Indiana':'IN', 'Iowa':'IA', 'Kansas':'KS', 'Kentucky':'KY', 'Louisiana':'LA', 'Maine':'ME', 'Maryland':'MD',
            'Massachusetts':'MA', 'Michigan':'MI', 'Minnesota':'MN', 'Mississippi':'MS', 'Missouri':'MO', 'Montana':'MT',
            'Nebraska':'NE', 'Nevada':'NV', 'New Hampshire':'NH', 'New Jersey':'NJ', 'New Mexico':'NM','New York':'NY', 'North Carolina':'NC',
            'North Dakota':'ND', 'Ohio':'OH', 'Oklahoma':'OK', 'Oregon':'OR', 'Pennsylvania':'PA', 'Rhode Island':'RI', 
            'South Carolina':'SC', 'South Dakota':'SD', 'Tennessee':'TN', 'Texas':'TX', 'Utah':'UT', 'Vermont':'VT',
            'Virginia':'VA', 'Washington':'WA', 'West Virginia':'WV', 'Wisconsin':'WI', 'Wyoming':'WY'
        }

    dataframe['state_po'] = dataframe['State'].map(state_abbr)
    dataframe.loc[dataframe['District'] == 'AL', 'District'] = 0
    dataframe = dataframe.rename(columns={'District':'district'})
    dataframe = dataframe.drop(['State', 'Party'], axis = 1)
    dataframe['year'] = year

    return dataframe

PVI2012 = processStateDistrict(PVI2012, 2012)
PVI2012['pvi'] = PVI2012['2012 Raw Cook PVI'].apply(process_pvi)
PVI2012 = PVI2012.drop(['2012 Winner', '2012 Cook PVI', '2012 Raw Cook PVI'], axis=1)
PVI2012
```

```python
def processStateDistrict22(dataframe, year):
    state_abbr = {
            'Alabama': 'AL', 'Alaska':'AK', 'Arizona':'AZ', 'Arkansas':'AR', 'California':'CA', 'Colorado':'CO', 
            'Connecticut':'CT','Delaware':'DE','Florida':'FL','Georgia':'GA', 'Hawaii':'HI','Idaho':'ID','Illinois':'IL',
            'Indiana':'IN', 'Iowa':'IA', 'Kansas':'KS', 'Kentucky':'KY', 'Louisiana':'LA', 'Maine':'ME', 'Maryland':'MD',
            'Massachusetts':'MA', 'Michigan':'MI', 'Minnesota':'MN', 'Mississippi':'MS', 'Missouri':'MO', 'Montana':'MT',
            'Nebraska':'NE', 'Nevada':'NV', 'New Hampshire':'NH', 'New Jersey':'NJ', 'New Mexico':'NM','New York':'NY', 'North Carolina':'NC',
            'North Dakota':'ND', 'Ohio':'OH', 'Oklahoma':'OK', 'Oregon':'OR', 'Pennsylvania':'PA', 'Rhode Island':'RI', 
            'South Carolina':'SC', 'South Dakota':'SD', 'Tennessee':'TN', 'Texas':'TX', 'Utah':'UT', 'Vermont':'VT',
            'Virginia':'VA', 'Washington':'WA', 'West Virginia':'WV', 'Wisconsin':'WI', 'Wyoming':'WY'
        }

    dataframe['state_po'] = dataframe['State'].map(state_abbr)
    dataframe.loc[dataframe['Number'] == 'AL', 'Number'] = 0
    dataframe = dataframe.rename(columns={'Number':'district'})
    dataframe = dataframe.drop(['State', 'Party'], axis = 1)
    dataframe['year'] = year

    return dataframe
```
